#### Measure of spread

```python
# Measures of Centre for numeric columns
num_cols = ["age", "hdlngth", "skullw", "totlngth", "taill", "footlgth", "earconch", "chest", "belly"]
```

```python
# checking for the mean
print("Mean:\n", vic_pop[num_cols].mean().round(2))
```

```python
# Checking for the median
print("Median:\n", vic_pop[num_cols].median().round(2))
```

```python
# checking for the mode
print("Mode:\n", vic_pop[num_cols].mode().iloc[0].round(2))
```

```python
# checking for the Variance
print("Variance\n", vic_pop[num_cols].var().round(2))
```

#### Measure of spread

```python
max = vic_pop[num_cols].max().round(2)
min = vic_pop[num_cols].min().round(2)

print("Maximum:", max)
print("\nMinimum:", min)
```
